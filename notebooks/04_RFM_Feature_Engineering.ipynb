# AI-Powered Customer Segmentation

## Notebook 04 : RFM Feature Engineering

### Objectives

- Create Recency
- Create Frequency
- Create Monetary
- Build Customer Level Dataset
- Save RFM Dataset

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns",None)
```

```python
df = pd.read_csv("/content/sample_data/eda_ready_data.csv")

df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df.head()
```

```python
print(df.shape)

df.info()
```

```python
print("Minimum Date :",df["InvoiceDate"].min())

print("Maximum Date :",df["InvoiceDate"].max())
```

```python
snapshot_date = df["InvoiceDate"].max() + pd.Timedelta(days=1)

print(snapshot_date)
```

```python
recency = (df.groupby("CustomerID")["InvoiceDate"].max())

recency = (snapshot_date-recency).dt.days
recency.head()
```

```python
frequency = (df.groupby("CustomerID")["InvoiceNo"].nunique())

frequency.head()
```

```python
monetary = (df.groupby("CustomerID")["TotalPrice"].sum())

monetary.head()
```

```python
rfm = pd.DataFrame({
    "Recency":recency,
    "Frequency":frequency,
    "Monetary":monetary
})

rfm.head()
```

```python
rfm = rfm.reset_index()
rfm.head()
```

```python
print(rfm.shape)
```

```python
rfm.isnull().sum()
```

```python
rfm.describe()
```

```python
rfm.sort_values("Monetary",ascending=False).head(10)
```

```python
rfm.sort_values("Frequency",ascending=False).head(10)
```

```python
rfm.sort_values("Recency").head(10)
```

```python
rfm.to_csv("/content/sample_data/rfm_data.csv",index=False)

print("RFM Dataset Saved Successfully")
```

# Business Insights

### Recency

- Lower Recency means customers purchased more recently.
- Higher Recency indicates inactive customers.

### Frequency

- Higher Frequency customers purchase more often.
- Low Frequency customers are occasional buyers.

### Monetary

- Higher Monetary customers contribute more revenue.
- Lower Monetary customers have lower business value.
